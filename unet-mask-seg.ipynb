```python
import os
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
from matplotlib.ticker import MaxNLocator
from PIL import Image

def plot_images(prediction_folder, actual_folder):
    # Dosya yollarını al
    prediction_files = os.listdir(prediction_folder)
    actual_files = os.listdir(actual_folder)

    # Dosya isimlerini sırala
    prediction_files.sort()
    actual_files.sort()

    # Her iki dosyadaki resimleri sırayla göster
    for prediction_file, actual_file in zip(prediction_files, actual_files):
        prediction_path = os.path.join(prediction_folder, prediction_file)
        actual_path = os.path.join(actual_folder, actual_file)

        # Resimleri aç
        prediction_img = Image.open(prediction_path)
        actual_img = Image.open(actual_path)

        # Subplot oluştur
        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

        # Prediction resmini göster
        axes[0].imshow(prediction_img)
        axes[0].set_title('Prediction')
        axes[0].axis('off')

        # Actual resmini göster
        axes[1].imshow(actual_img)
        axes[1].set_title('Actual')
        axes[1].axis('off')

        # Subplot'ları göster
        plt.show()

# Prediction ve Actual dosya yollarını belirt
prediction_folder_path = '/home/Desktop/my/xray-segmentatation/U-2-Net/test_data/test_images_xray _results'
actual_folder_path = '/home/Desktop/my/xray-segmentatation/U-2-Net/test_data/mask'

# Fonksiyonu çağır
plot_images(prediction_folder_path, actual_folder_path)
```
